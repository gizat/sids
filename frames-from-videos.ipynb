```
import cv2
import numpy as np


folder_path = '/content/drive/My Drive/Masters/Colab Notebooks/Research Code/data/'
folder = 'positive-night/'

# Read the video file
cap = cv2.VideoCapture(folder_path + folder + 'video.mp4')
fps = 1                     # frames per seconds
total_frames = 70 * fps     # total frames to read in N seconds
counter = 0

# Read a frame from the video file. 
# Return boolean if success or not
ret, frame = cap.read()

print(ret)

while ret:
    cv2.imwrite(folder_path + folder + 'frames-1m/' + 'frame%d.jpg' % counter, frame)     # save frame as JPEG file
    
    ret, frame = cap.read()     # repeat the process
    print('Read a new frame: ', counter)
    
    counter += 1

    if counter == total_frames:
        ret = False

# When everything done, release the capture
#cap.release()
#cv2.destroyAllWindows()
```
